```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("booksc.csv", encoding='ISO-8859-1')
print("游댳 First 5 Rows of Data:\n", df.head(), "\n")
print("游댳 Dataset Shape:", df.shape, "\n")
print("游댳 Data Types:\n", df.dtypes, "\n")
print("游댳 Unique Values per Column:\n", df.nunique(), "\n")
print("游댳 Descriptive Statistics:\n", df.describe(include='all'), "\n")
df['Price'] = df['Price'].replace(r'[^\d.]', '', regex=True).astype(float)

print("游댳 Missing Values:\n", df.isnull().sum(), "\n")

plt.figure(figsize=(10, 6))
sns.histplot(df['Price'], bins=30, kde=True)
plt.title("Distribution of Book Prices")
plt.xlabel("Price")
plt.ylabel("Count")
plt.show()

if 'Category' in df.columns:
    plt.figure(figsize=(10, 6))
    sns.countplot(data=df, y='Category', order=df['Category'].value_counts().index)
    plt.title("Top Book Categories")
    plt.xlabel("Count")
    plt.ylabel("Category")
    plt.show()
```
